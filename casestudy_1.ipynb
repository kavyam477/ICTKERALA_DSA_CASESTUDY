```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### 1. Read the dataset to the python environment

```python
data=pd.read_csv("cars_data.csv")
```

```python
data
```

### 2. Check for the null values present in the dataset

```python
data.isna()
```

### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure()
slice_data1=data[(data["Buyer Gender"]=="Female")]
a=len(slice_data1)
slice_data2=data[(data["Buyer Gender"]=="Male")]
b=len(slice_data2)
print("No of females :",a)
print("No of Males :",b)
x=["Female","Male"]
y=[a,b]
sns.barplot(x,y)
plt.show()
```

### 4. Find the top 5 cars based on their sales price.

```python
sp = data.sort_values(by = 'Sale Price',ascending= False)
```

```python
sp1=sp.head()
sp1
```

```python
plt.figure(figsize=(10,10))
x=sp1['Model']
y=sp1['Sale Price']
plt.xlabel('Model')
plt.ylabel('Sale Price')
plt.title("Top 5 cars based on their sales price")
plt.barh(x,y)
for index, value in enumerate(y):
    plt.text(value, index,
      str(value))
plt.show()
```

### 5. Find the least 5 cars based on their Resell price. 

```python
rp = data.sort_values(by = 'Resell Price',ascending= False)
```

```python
rp1=rp.tail()
rp1
```

```python
plt.figure(figsize=(10,10))
x=rp1['Model']
y=rp1['Resell Price']
plt.xlabel('Resell Price')
plt.ylabel('Model')
plt.title("Least 5 cars based on their Reselling Price")
plt.barh(x,y)
for index, value in enumerate(y):
    plt.text(value, index,
             str(value))
plt.show()
```
