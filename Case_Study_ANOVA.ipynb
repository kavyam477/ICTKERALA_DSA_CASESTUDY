### 1. The mean sales generated by each zone. 

```python
print("Means generated by Zone-A : ",data1['Zone - A'].mean())
print("Means generated by Zone-B : ",data1['Zone - B'].mean())
print("Means generated by Zone-C : ",data1['Zone - C'].mean())
print("Means generated by Zone-D : ",data1['Zone - D'].mean())
```

### 2. Total sales generated by all the zones for each month.

```python
data1['Total']=sum((data1['Zone - A'],data1['Zone - B'],data1['Zone - C'],data1['Zone - D']))
data1
```

### 3.  Check whether all the zones generate the same amount of sales.

```python
import scipy.stats as stats
fvalue,pvalue = stats.f_oneway(data1['Zone - A'],data1['Zone - B'],data1['Zone - C'],data1['Zone - D'])
print("fvalue is : ",fvalue)
print("pvalue is : ",pvalue)
```

```python
if(pvalue < 0.05):
    print("P value is less than 0.05")
    print("Null Hypothesis is rejected.")
    print("Amount of sales generated by the four zones are not same. Atleast one zone has generated a different amount of sales")
else:
    print("P value is greater than 0.05")
    print("Null Hypothesis is accepted.")
    print("All Zones generated same amount of Sales")
```
